```python
import pandas as pd
import numpy as np
from numpy import nan as NA
```

```python
df1 = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
df1
```

### 1. Some values in the the FlightNumber column are missing. These numbers are meant to increase by 10 with  each row so 10055 and 10075 need to be put in  place.Fill in these missing numbers and make the column an integer column(instead of a float column)
      

```python
df1['FlightNumber'] = df1['FlightNumber'].interpolate().astype(int)
df1
```

### 2.The From_To column would be better as two separate columns! Split each string on the underscore delimeter _to give a new temperory DataFrame with the correct values.Assign the correct column names to this temperory DataFrame

```python
temp=pd.DataFrame()
temp['From']=df1['From_To'].str.split('_')
temp['To']=df1['From_To'].str.split('_')
temp
```

### 3. Notice how the capitalisation of the city names is all mixed up in this temperory DataFrame.Standardise the strings so that only the first letter is uppercase(e.g "londoN" should become "London")


```python
temp['From']=df1['From_To'].str.split().str[0].str.capitalize()
temp['To']=df1['From_To'].str.split('_').str[1].str.capitalize()
temp
```

### 4. Delete the From_To column from df and attach the temporary DataFrame from the previous questions

```python
df1.pop('From_To')
df1
```

```python
df1=pd.concat([df1,temp[['From','To']]],axis=1)
df1
```
